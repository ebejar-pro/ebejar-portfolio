```python
import sqlite3
import pandas as pd
from pathlib import Path

# Connect to SQLite database (creates file if it doesn't exist)
conn = sqlite3.connect("studentgrades.db")

cursor = conn.cursor()
 # Create company employees table
cursor.execute('''
        CREATE TABLE IF NOT EXISTS studentgrades (
            student_id INTEGER PRIMARY KEY,
            subject TEXT,
            score INTEGER,
            grade TEXT
        )
        ''')
conn.commit()
conn.close()

# Load data from CSV into DataFrame
# Paths
BASE_DIR = Path.cwd().parent
CSV_PATH = BASE_DIR / "data" / "raw" / "student_grades.csv"
DB_PATH  = BASE_DIR / "data" / "studentgrades.db"

# Check CSV Path exists
if not CSV_PATH.exists():
    raise FileNotFoundError(f"CSV not found at: {CSV_PATH}")

# Load CSV
df = pd.read_csv(CSV_PATH)

# Connect to SQLite
conn = sqlite3.connect(DB_PATH)

# Insert data
df.to_sql(
    name="studentgrades",
    con=conn,
    if_exists="append",
    index=False
)

# Close connection
conn.commit()
conn.close()

print("✅ Data successfully inserted into studentgrades table.")
```
