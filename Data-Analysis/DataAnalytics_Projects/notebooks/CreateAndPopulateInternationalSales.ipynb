```python
import sqlite3
import os

# Connect to SQLite database (creates file if it doesn't exist)
conn = sqlite3.connect("inter_sales.db")

cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS inter_sales (
    sales_id INTEGER PRIMARY KEY,
    sales_person TEXT NOT NULL,
    country TEXT NOT NULL,
    sale_amount REAL NOT NULL,
    sale_date DATE NOT NULL
);
""")

print("Current working directory:", os.getcwd())

conn.commit()
conn.close()
```

```python
import sqlite3
import pandas as pd
from pathlib import Path


BASE_DIR = Path(
    r"C:\Users\elbej\OneDrive\Documents\GitHub\ebejar-portfolio\Data-Analysis\DataAnalytics_Projects"
)

##BASE_DIR = Path.cwd()          # <-- Jupyter-safe

CSV_PATH = BASE_DIR / "data" / "raw" / "international_sales.csv"
DB_PATH  = BASE_DIR / "data" / "inter_sales.db"

print("Using BASE_DIR:", BASE_DIR)
print("CSV_PATH:", CSV_PATH)
print("DB_PATH:", DB_PATH)



print("Using DB:", DB_PATH.resolve())

df = pd.read_csv(CSV_PATH)

assert df.isnull().sum().sum() == 0, "Null values found!"
assert (df["sale_amount"] >= 0).all(), "Invalid sale_amount!"

df["sale_date"] = pd.to_datetime(df["sale_date"]).astype(str)

conn = sqlite3.connect(DB_PATH)

df.to_sql(
    name="inter_sales",
    con=conn,
    if_exists="append",
    index=False
)

conn.commit()
conn.close()

print("✅ Data successfully inserted into inter_sales.")
```
