```python
import sqlite3
import random
from datetime import datetime, timedelta

# 1. Connect to your database (Use your working full path here)
db_path = r'C:\Users\elbej\OneDrive\Documents\GitHub\ebejar-portfolio\Data-Analysis\DataAnalytics_Projects\data\sales.db' 
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# 2. Create the Sales table with sales_id as Primary Key
cursor.execute('''
CREATE TABLE IF NOT EXISTS sales (
    sales_id INTEGER PRIMARY KEY AUTOINCREMENT,
    sales_person TEXT NOT NULL,
    sale_amount REAL NOT NULL,
    sale_date TEXT NOT NULL
)
''')

# 3. Fetch current employees to ensure names match
cursor.execute("SELECT sales_person FROM sales")
employees = [row[0] for row in cursor.fetchall()]

# 4. Generate Synthetic Data
sales_data = []

for person in employees:
    for month in range(1, 13):  # Months 1 to 12
        for _ in range(5):      # 5 sales per person per month
            # Generate a random day (1-28 to avoid February issues)
            day = random.randint(1, 28)
            # Generate random time
            hour = random.randint(8, 20)
            minute = random.randint(0, 59)
            
            sale_date = f"2023-{month:02d}-{day:02d} {hour:02d}:{minute:02d}:00"
            # Random amount between 100.00 and 5000.00
            amount = round(random.uniform(100, 5000), 2)
            
            sales_data.append((person, amount, sale_date))

# 5. Insert data into the table
# We don't include sales_id because it is AUTOINCREMENT
cursor.executemany('''
    INSERT INTO sales (sales_person, sale_amount, sale_date) 
    VALUES (?, ?, ?)
''', sales_data)

conn.commit()
print(f"Successfully inserted {len(sales_data)} records into the sales table.")

# Close connection
conn.close()
```
