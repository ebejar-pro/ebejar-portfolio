```python
import sqlite3
import pandas as pd
from pathlib import Path

# Connect to SQLite database (creates file if it doesn't exist)
conn = sqlite3.connect("climbers.db")

cursor = conn.cursor()
 # Create climbers table
cursor.execute('''
        CREATE TABLE IF NOT EXISTS climbers (
            id INTEGER PRIMARY KEY,
            climbers_name TEXT,
            nationality TEXT,
            successful_climbs INTEGER,
            failed_climbs INTEGER,
            date_of_first_ascent DATE,
            date_of_last_ascent DATE
        )
        ''')
conn.commit()
conn.close()

# Load data from CSV into DataFrame
# Paths
BASE_DIR = Path.cwd().parent
CSV_PATH = BASE_DIR / "data" / "raw" / "climbers.csv"
DB_PATH  = BASE_DIR / "data" / "climbers.db"

# Check CSV Path exists
if not CSV_PATH.exists():
    raise FileNotFoundError(f"CSV not found at: {CSV_PATH}")

# Load CSV
df = pd.read_csv(CSV_PATH)

# Connect to SQLite
conn = sqlite3.connect(DB_PATH)

# Insert data
df.to_sql(
    name="climbers",
    con=conn,
    if_exists="append",
    index=False
)

# Close connection
conn.commit()
conn.close()

print("✅ Data successfully inserted into climbers table.")
```
