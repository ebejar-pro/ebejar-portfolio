```python
import sqlite3
import os

# Connect to SQLite database (creates file if it doesn't exist)
conn = sqlite3.connect("coffee_sales.db")

cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS coffee_sales (
    coffee_sale_id INTEGER PRIMARY KEY,
    date TEXT NOT NULL,
    datetime TEXT NOT NULL,
    cash_type TEXT NOT NULL CHECK (cash_type IN ('cash', 'card')),
    card TEXT,
    money REAL NOT NULL CHECK (money >= 0),
    coffee_name TEXT NOT NULL
);

""")

print("Current working directory:", os.getcwd())

conn.commit()
conn.close()
```

```python
import sqlite3
import pandas as pd
from pathlib import Path


BASE_DIR = Path(
    r"C:\Users\elbej\OneDrive\Documents\GitHub\ebejar-portfolio\Data-Analysis\DataAnalytics_Projects"
)

##BASE_DIR = Path.cwd()          # <-- Jupyter-safe

CSV_PATH = BASE_DIR / "data" / "raw" / "coffee_sales.csv"
DB_PATH  = BASE_DIR / "data" / "coffee_sales.db"

print("Using BASE_DIR:", BASE_DIR)
print("CSV_PATH:", CSV_PATH)
print("DB_PATH:", DB_PATH)



print("Using DB:", DB_PATH.resolve())

df = pd.read_csv(CSV_PATH)

conn = sqlite3.connect(DB_PATH)

df.to_sql(
    name="coffee_sales",
    con=conn,
    if_exists="append",
    index=False
)

conn.commit()
conn.close()

print("✅ Data successfully inserted into coffee_sales.")
```
