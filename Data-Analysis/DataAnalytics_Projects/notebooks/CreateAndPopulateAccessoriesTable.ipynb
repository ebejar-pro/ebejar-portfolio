```python
import sqlite3
import pandas as pd

def add_data_to_database():
    try:
        conn = sqlite3.connect('accessories.db')
        cursor = conn.cursor()
        # Create accessories table
        cursor.execute('''
        CREATE TABLE IF NOT EXISTS accessories (
            accessory_id INTEGER PRIMARY KEY,
            accessory_name TEXT,
            parent_id INTEGER
        )
        ''')
        # Sample data for accessories table
        accessories_data = [
            (1, 'Electronics', None),
            (2, 'Computers', 1 ),
            (3, 'Laptops', 2),      
            (4, 'Desktops', 2),
            (5, 'Mobile Phones', 1),
            (6, 'Accessories', 5),
            (7, 'Cases', 6),
            (8, 'Headphones', 6)        
            ]   
        cursor.executemany(
            '''
            INSERT OR REPLACE INTO accessories (accessory_id, accessory_name, parent_id)
            VALUES (?, ?, ?)
            ''', accessories_data)
        conn.commit()
    except Exception as e:
        print(f"An error occurred: {e}")    
    finally:
        conn.close()    
# Calling message
add_data_to_database()
```
