```python
import sqlite3
import pandas as pd

def add_data_to_database():
    try:
        conn = sqlite3.connect('workrecords.db')
        cursor = conn.cursor()
        # Create accessories table
        cursor.execute('''
        CREATE TABLE IF NOT EXISTS workrecords (
            id INTEGER PRIMARY KEY,
            first_name TEXT,
            days_worked INTEGER,
            earnings INTEGER,
            team_id INTEGER
        )
        ''')
        # Sample data for workrecords table
        workrecords_data = [
            (1, 'John', 10, 2000, 101),
            (2, 'Mary', 15, 2500, 102),
            (3, 'Alice', 20, 3000, 103),      
            (4, 'Alice', 18, 1700, 103),
            (5, 'Mary', 26, 4000, 102),
            (6, 'John', 17, 2600, 101)       
            ]   
        cursor.executemany(
            '''
            INSERT OR REPLACE INTO workrecords (id, first_name, days_worked, earnings, team_id)
            VALUES (?, ?, ?, ?, ?)
            ''', workrecords_data)
        conn.commit()
    except Exception as e:
        print(f"An error occurred: {e}")    
    finally:
        conn.close()    
# Calling message
add_data_to_database()
```
