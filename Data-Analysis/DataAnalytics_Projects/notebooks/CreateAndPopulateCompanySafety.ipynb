```python
import sqlite3
import pandas as pd
from pathlib import Path

# Connect to SQLite database (creates file if it doesn't exist)
conn = sqlite3.connect("safety.db")

cursor = conn.cursor()
 # Create company safety table
cursor.execute('''
        CREATE TABLE IF NOT EXISTS safety (
            employee_id INTEGER PRIMARY KEY,
            employee_name TEXT,
            department TEXT,
            incident_date DATE,
            incident_type TEXT,
            day_lost INTEGER,
            injury_severity TEXT
        )
        ''')
conn.commit()
conn.close()

# Load data from CSV into DataFrame
# Paths
BASE_DIR = Path.cwd().parent
CSV_PATH = BASE_DIR / "data" / "raw" / "company_safety.csv"
DB_PATH  = BASE_DIR / "data" / "safety.db"

# Check CSV Path exists
if not CSV_PATH.exists():
    raise FileNotFoundError(f"CSV not found at: {CSV_PATH}")

# Load CSV
df = pd.read_csv(CSV_PATH)

# Connect to SQLite
conn = sqlite3.connect(DB_PATH)

# Insert data
df.to_sql(
    name="safety",
    con=conn,
    if_exists="append",
    index=False
)

# Close connection
conn.commit()
conn.close()

print("✅ Data successfully inserted into safety table.")
```
