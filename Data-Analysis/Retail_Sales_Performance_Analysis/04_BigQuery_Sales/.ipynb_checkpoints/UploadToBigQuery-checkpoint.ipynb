```python
import pandas as pd
from faker import Faker
from google.cloud import bigquery
import os


# 1. AUTHENTICATION (Replace with your JSON key path)
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "C:/Users/elbej/bq-analysis-04-9ce735113ac2.json"

# 2. GENERATE SYNTHETIC RETAIL DATA
fake = Faker()
data = []
for _ in range(100):  # Generating 100 fake transactions
    data.append({
        "transaction_id": fake.uuid4(),
        "store_id": fake.random_int(min=1, max=20),
        "customer_id": f"CUST-{fake.random_int(1000, 9999)}",
        "transaction_timestamp": fake.date_time_this_month().isoformat(),
        "total_amount": round(fake.pyfloat(left_digits=3, right_digits=2, min_value=10, max_value=500), 2)
    })

df = pd.DataFrame(data)

# 3. UPLOAD TO BIGQUERY
client = bigquery.Client()
table_id = "bq-analysis-04.retail_bronze.raw_sales" # Dataset must exist first

job_config = bigquery.LoadJobConfig(
    write_disposition="WRITE_APPEND", # Adds new data to the bottom
    autodetect=True                   # Automatically detects schema/types
)

print(f"Uploading {len(df)} rows to {table_id}...")
job = client.load_table_from_dataframe(df, table_id, job_config=job_config)
job.result() # Wait for job to finish

print("✅ Success! Check your BigQuery console.")
```
