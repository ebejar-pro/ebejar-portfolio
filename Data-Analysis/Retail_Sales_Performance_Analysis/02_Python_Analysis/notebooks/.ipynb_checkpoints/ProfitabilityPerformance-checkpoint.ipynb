```python
import pandas as pd
import matplotlib.pyplot as plt

# -----------------------------
# 1. Sample Employees Table
# -----------------------------
employees_data = {
    "employee_id": [1, 2, 3],
    "name": ["Alice Smith", "Bob Lee", "Rolly Mars"],
    "salary": [50000, 48000, 45000],
    "job_code": ["SALES", "SALES", "SALES"]
}
employees = pd.DataFrame(employees_data)

# -----------------------------
# 2. Sample Products Table
# -----------------------------
products_data = {
    "product_id": [1, 2, 3],
    "product_name": ["Laptop", "Phone", "Tablet"],
    "price": [1200.00, 800.00, 400.00],
    "cost_price": [700.00, 450.00, 250.00]
}
products = pd.DataFrame(products_data)

# -----------------------------
# 3. Sample Sales Table
# -----------------------------
sales_data = {
    "sale_id": [101, 102, 103, 104],
    "employee_id": [1, 2, 3, 3],
    "product_id": [1, 2, 3, 1],
    "quantity": [2, 3, 5, 1],
    "sale_date": pd.to_datetime(["2025-09-12", "2025-09-13", "2025-09-14", "2025-09-15"])
}
sales = pd.DataFrame(sales_data)

# -----------------------------
# 4. Sample Sales Commission Table
# -----------------------------
commission_data = {
    "commission_id": [1, 2, 3, 4],
    "employee_id": [1, 2, 3, 3],
    "sale_id": [101, 102, 103, 104],
    "commission_amount": [200.00, 160.00, 300.00, 120.00],
    "created_at": pd.to_datetime(["2025-09-12", "2025-09-13", "2025-09-14", "2025-09-15"])
}
sales_commission = pd.DataFrame(commission_data)

# -----------------------------
# 5. Calculate Total Sales per Employee
# -----------------------------
sales_with_products = sales.merge(products, on="product_id", how="left")
sales_with_products["sales_value"] = sales_with_products["quantity"] * sales_with_products["price"]

total_sales = sales_with_products.groupby("employee_id")["sales_value"].sum().reset_index()

# -----------------------------
# 6. Calculate Total Commission per Employee
# -----------------------------
total_commission = sales_commission.groupby("employee_id")["commission_amount"].sum().reset_index()

# -----------------------------
# 7. Merge with Employees Table
# -----------------------------
performance = employees.merge(total_sales, on="employee_id", how="left")
performance = performance.merge(total_commission, on="employee_id", how="left")

# Fill missing values with 0
performance[["sales_value", "commission_amount"]] = performance[["sales_value", "commission_amount"]].fillna(0)

print(performance)

# -----------------------------
# 8. Visualization
# -----------------------------
plt.figure(figsize=(10, 6))
bar_width = 0.25
x = range(len(performance))

plt.bar([i - bar_width for i in x], performance["salary"], width=bar_width, label="Salary", color="steelblue")
plt.bar(x, performance["commission_amount"], width=bar_width, label="Commission", color="orange")
plt.bar([i + bar_width for i in x], performance["sales_value"], width=bar_width, label="Sales Value", color="green")

plt.xticks(x, performance["name"], rotation=45)
plt.title("Employee Performance: Salary vs Commission vs Sales")
plt.ylabel("Amount ($)")
plt.legend()
plt.tight_layout()
plt.show()
```
