```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns

# Load Data
sales = pd.read_csv('../data/sales.csv')
customers = pd.read_csv('../data/customers.csv')
products = pd.read_csv('../data/products.csv')

# Merge sales with products to get price
sales = sales.merge(
    products[['product_id', 'price']],
    on='product_id',
    how='left'
)

# Revenue calculation
sales['revenue'] = sales['price'] * sales['quantity']

# Aggregate customer-level KPIs
cust = sales.groupby('customer_id').agg({
    'revenue': 'sum',
    'quantity': 'sum',
    'sale_id': 'count'
}).reset_index()

cust.rename(columns={
    'sale_id': 'frequency',
    'quantity': 'units_bought'
}, inplace=True)

# Create age groups
customers['age_group'] = pd.cut(
    customers['age'],
    bins=[0, 25, 35, 50, 100],
    labels=['1825', '2635', '3650', '50+']
)

# Merge customer demographics
cust = cust.merge(customers[['customer_id', 'age', 'age_group']], on='customer_id', how='left')

# Feature selection
X = cust[['revenue', 'frequency', 'units_bought', 'age']]

# Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Elbow Method
inertias = []
for k in range(1,10):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    inertias.append(km.inertia_)

plt.plot(range(1,10), inertias, marker='o')
plt.title("Elbow Method")
plt.xlabel("Clusters")
plt.ylabel("Inertia")
plt.show()

# Choose 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)
cust['segment'] = kmeans.fit_predict(X_scaled)

# Segment summary
segment_summary = cust.groupby('segment').agg({
    'revenue': 'mean',
    'frequency': 'mean',
    'units_bought': 'mean',
    'age': 'mean',
    'customer_id': 'count'
})
print(segment_summary)

# Visualization
plt.figure(figsize=(10,6))
sns.scatterplot(data=cust, x='revenue', y='frequency', hue='segment', palette='tab10')
plt.title("Customer Segments")
plt.show()

# Export segmented customers
cust.to_csv('../data/customer_segments.csv', index=False)
```
